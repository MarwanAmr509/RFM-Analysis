```python
# import modules
import pandas as pd # for dataframes
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# reading dataset
data = pd.read_csv("C:\\Users\\marwan\\Downloads\\Online Retail.csv")
df=pd.DataFrame(data)
```

```python
# first 5 rows
df.head()
```

```python
data.shape
```

```python
# Exploring the different regions of transactions
data.Country.unique()
```

```python
# Stripping extra spaces in the description
df['Description'] = df['Description'].str.strip()
```

```python
df['Description'].isnull().sum()
```

```python
# Dropping the rows without any invoice number
data.dropna(axis = 0, subset =['InvoiceNo'], inplace = True)
```

```python
#convert data in InvoiceNo to string
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
```

```python
# Dropping all row that contain null values if Description is null
df.dropna(axis=0, subset=['Description'], inplace=True)
```

```python
# Dropping all transactions which were done on credit
data = data[~data['InvoiceNo'].str.contains('C')]
```

```python
#filter
#Splitting the data according to the United Kingdom
# Transactions done in United Kingdom
basket_UK = (df[df['Country'] =="United Kingdom"]
.groupby(['InvoiceNo', 'Description'])['Quantity']
.sum().unstack().reset_index().fillna(0)
.set_index('InvoiceNo'))

basket_UK.head()
```

```python
#convert all cell to 0 and 1
def unit_encode(x):

    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
basket_UK_sets=basket_UK.applymap(unit_encode)
basket_UK_sets.drop("POSTAGE",inplace=True,axis=1)
```

```python
basket_UK=basket_UK_sets
```

```python
frequent_itemsets = apriori(basket_UK, min_support=0.03, use_colnames=True)
```

```python
frequent_itemsets.head()
```

```python
# lift
# Collecting the inferred rules in a dataframe
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules=rules.sort_values(['confidence','lift'],ascending=[False,False])
rules.head()
```

```python
print(rules[(rules['lift']>=6) & (rules['confidence']>=0.6)])
```
